#     *c. Encoding*

```python
data_trainloan2.info()
```

```python
data_trainloan3 = data_trainloan2
```

```python
data_trainloan3.info()
#Using data_trainloan3 from here onwards for encoding
```

```python
data_trainloan3.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
data_trainloan3['Dependents'].unique()
```

```python
# Label encoding column 'Dependents'
data_trainloan3['Dependents'] = le.fit_transform(data_trainloan3['Dependents'])
data_trainloan3['Dependents'].unique()
```

```python
# One hot encoding columns 'Gender','Married','Education','Self_Employed','Property_Area','Loan_Status'
data_trainloan3_encoded = pd.get_dummies(data_trainloan3, columns= ['Gender','Married','Education','Self_Employed','Property_Area','Loan_Status'], dtype = int)
```

```python
data_trainloan3_encoded.head(10)
```

```python
data_trainloan3_encoded.info()
```

#     *d. Scaling*

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
data_x = data_trainloan3_encoded.drop('Loan_ID',axis = 1)
```

```python
num_cols = ['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']
data_x.loc[:,num_cols] = sc.fit_transform(data_x[num_cols])
```
